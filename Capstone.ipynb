```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataset = pd.read_excel("Netflix.xlsx")
Dataset.head()
```

```python
Dataset.info()
```

###  Note : Here from the above infor function we can see that most of the columns are given as object and also most f the columns contains null values.

```python
Dataset.shape
```

```python
null_values = Dataset.isnull().sum().reset_index()
null_values
```

#### here we could see that there are many columns with nulll values, 

```python
nuls1 = null_values[null_values[0] < 1000]
nuls1
```

#### I have filtered the null values which are less than thousand and saved them to nuls1

```python
Dataset.head().T
```

```python
for i in nuls1['index']:
    Dataset.dropna(subset=[i], inplace=True)
```

#### here i have dropped the all the null values which are less than 1000

```python
Dataset.isnull().sum()
```

#### Here we could see there are few moer null values 

```python
Dataset['Rotten Tomatoes Score'].fillna(value=Dataset['Rotten Tomatoes Score'].mean(), inplace=True) 
```

```python
Dataset['Metacritic Score'].fillna(value=58.019707, inplace=True)
Dataset['Awards Received'].fillna(value=Dataset['Awards Received'].median(), inplace=True) 
Dataset['Awards Nominated For'].fillna(value=Dataset['Awards Nominated For'].median(), inplace=True) 
Dataset['Boxoffice'].fillna(value=Dataset['Boxoffice'].mean(), inplace=True) 
```

#### These are few of the numericals columns which are having the null values I have tried to 
#### fill the null values by using the mean() , meadian().

```python
Dataset.isnull().sum()
```

```python
Dataset['Director'].fillna('unknown',inplace=True)
Dataset['Writer'].fillna('unknown',inplace=True)
Dataset['View Rating'].fillna('unknown',inplace=True)
Dataset['Production House'].fillna('unknown',inplace=True)
```

#### Here I am trying to deal with the categorical null values here is can use mode to fill the values because there are almost 
#### 2000 to 3000 values which are nullso i am trying to fill them with unknown if I fill mode the can become as a false data.

##### i.e: If I fill with mode the missing 3000 or 4000 null values will be filled with a same name.
